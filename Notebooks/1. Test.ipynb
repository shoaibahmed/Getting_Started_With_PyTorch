```python
i = torch.LongTensor([[0, 2], [1, 0], [1, 2]])
v = torch.FloatTensor([3, 4, 5])
```

```python
torch.sparse.FloatTensor(i.t(), v, torch.Size([2,3])).to_dense()
```

```python
import numpy as np
```

```python
i = np.array([[0, 2], [1, 0], [1, 2]])
v = np.array([3, 4, 5])
```

```python
import scipy.sparse
```

```python
sp = scipy.sparse.csr_matrix((v, (i[:, 0], i[:, 1])), shape=(3,3))
```

```python
sp.todense()
```

```python
sp
```

```python
sp[1,:]
```

```python
sp[1,:].todense()
```

```python
sp[:, 1].todense()
```

```python
np.array(sp[:, 1].todense()).flatten()
```
